```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```
import pandas_datareader as web
dataset = web.DataReader("BANKINDIA.NS",data_source = "yahoo", start = "2000-01-01",end = "2020-06-30 ")
```

```
plt.figure(figsize=(30,10))
plt.plot(dataset["Open"],color ='blue')
sns.set_style('darkgrid')
```

```
plt.figure(figsize=(30,10))
sns.pairplot(dataset)
```

```
plt.figure(figsize=(30,10))
sns.heatmap(dataset.corr(),annot = True)
```

###SIMPLE MOVING AVERAGE ###

```
rolling_mean_9   = dataset.Open.rolling(window=9).mean()
rolling_mean_21  = dataset.Open.rolling(window=21).mean()
rolling_mean_25  = dataset.Open.rolling(window=25).mean()
rolling_mean_50  = dataset.Open.rolling(window=50).mean()
rolling_mean_100 = dataset.Open.rolling(window=100).mean()
rolling_mean_200 = dataset.Open.rolling(window=200).mean()
```

```
plt.figure(figsize=(30,10))
plt.plot( dataset.Open, label='BANK OF INDIA')
plt.plot( rolling_mean_50, label='50 Day SMA', color='green')
plt.plot( rolling_mean_100, label='100 Day SMA', color='red')
plt.legend()
plt.show()
```

```
plt.figure(figsize=(30,10))
plt.plot( dataset.Open, label='BANK OF INDIA')
plt.plot( rolling_mean_100, label='100 Day SMA', color='green')
plt.plot( rolling_mean_200, label='200 Day SMA', color='red')
plt.legend()
plt.show()
```
